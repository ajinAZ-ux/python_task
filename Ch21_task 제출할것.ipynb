```python
def odd_numbers(n):
    # 0부터 n까지 반복
    for i in range(n + 1):
        # i가 홀수면
        if i % 2 == 1:
            yield i   # i를 하나 내보내기
```

[문제 10] 팩토리얼을 계산하는 재귀 함수 factorial을 작성하세요.

```python
def factorial(n):
    if n == 0:
        return 1
    return n * factorial(n - 1)

print(factorial(5))
```

[문제 12] 두 문자열을 받아서 긴 문자열을 반환하는 함수 longer_string을 작성하세요. 만약 두 문자열의 길이가 같다면, 첫 번째 문자열을 반환하세요.

```python
def longer_string(a, b):
    if len(a) >= len(b):
        return a
    return b

print(longer_string("hello", "bye"))
```
